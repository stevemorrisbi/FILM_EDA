```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# reading in the data
df = pd.read_csv(r'C:\Users\smorr\Desktop\CV portfolio etc\2023 - New practical projects\Python\Film EDA\movies.csv') 
```

```python
df
```

```python
# removing duplicate entries from the df
df.drop_duplicates()
```

```python
# ordering by highest grossing film 
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Group the data by genre and calculate the average score and count of films for each genre
genre_avg_scores = df.groupby('genre')['score'].agg(['mean', 'count'])

# Sort the results by the average score in descending order
genre_avg_scores_sorted = genre_avg_scores.sort_values(by='mean', ascending=False)

# Print the result
print(genre_avg_scores_sorted)
```

```python
# Rename the 'count' column to 'total films'
genre_avg_scores.rename(columns={'count': 'total films'}, inplace=True)

# Filter genres where the count is >= 10
genre_avg_scores_filtered = genre_avg_scores[genre_avg_scores['total films'] >= 10]

# Sort the filtered results by the average score in descending order
genre_avg_scores_sorted = genre_avg_scores_filtered.sort_values(by='mean', ascending=False)

# Print the filtered and sorted result
print(genre_avg_scores_sorted)
```

```python
# renaming df
movies_data = df

# Define the list of genres you want to analyze
genres_to_analyze = ['Biography', 'Animation', 'Drama', 'Crime']

# Filter the data to include only the specified genres
filtered_data = movies_data[movies_data['genre'].isin(genres_to_analyze)]

# Group the filtered data by year and genre, and calculate the average score for each group
yearly_genre_avg_scores = filtered_data.groupby(['year', 'genre'])['score'].mean().unstack()

# Create a line chart to show the trend of average scores over the years for each genre
plt.figure(figsize=(12, 8))

# Set the x-axis to be the years
years = yearly_genre_avg_scores.index

for genre in genres_to_analyze:
    plt.plot(years, yearly_genre_avg_scores[genre], label=genre)

plt.title('Average Scores Over the Years by Genre')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5) <1.0 is to exclude exact matches 

strong_pairs = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1.0)]

print(strong_pairs)
```
